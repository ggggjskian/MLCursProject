#Приложение 3
Работа над представлением записи для передачи в модель


Представление записи

```python
import mne
import numpy as np
from scipy.signal import welch
import neurokit2 as nk





FREQ_BANDS = {
    'delta': [0.5, 4], 'theta': [4, 8], 'alpha': [8, 13],
    'beta': [13, 30], 'gamma': [30, 70]
}



def splitRecordOn(fileName : str):
  ECGNAME = "ECG V3-V1"
  rawData = mne.io.read_raw_edf(fileName, preload = True, verbose = "WARNING")
  dataEEG = rawData.copy()
  if ECGNAME in dataEEG.ch_names:
    dataEEG.drop_channels([ECGNAME])
  dataECG = rawData.copy().pick(ECGNAME)

  return filterRecordComponent(dataEEG, dataECG)

def filterRecordComponent(eeg, ecg):
    if eeg:
      eeg.filter(l_freq=0.5, h_freq=70, verbose='WARNING')
      eeg.notch_filter(freqs=50, verbose='WARNING')
    if ecg: ecg.filter(l_freq=5, h_freq=35, verbose='WARNING')
    return eeg, ecg

def splitRecordsOnDeltaT(fileName : str, dt : int): # параметр dt  в секундах
  dataEEG, dataECG = splitRecordOn(fileName)
  combineFilterData = dataEEG.copy()
  combineFilterData.add_channels([dataECG], force_update_info=True)

  events = mne.make_fixed_length_events(combineFilterData, duration= dt)
  epochs = mne.Epochs(combineFilterData, events, tmin = 0, tmax = dt, baseline=None, preload=True, verbose='WARNING')

  return epochs


def extractParametrsMatrix(epoch, sfreq = 1000):
    eeg_data = epoch[:-1, :]
    ecg_data = epoch[-1, :]

    features = []
    for channel in eeg_data:
      freqs, psd = welch(channel, fs=sfreq, nperseg=sfreq)
      for band, (fmin, fmax) in FREQ_BANDS.items():
          idx_band = np.logical_and(freqs >= fmin, freqs <= fmax)
          band_power = np.sum(psd[idx_band])
          features.append(band_power)


    try: #тут неободимо обработка, если алгоритм не сможет найти пики для промежутка
      _, rpeaks = nk.ecg_peaks(ecg_data, sampling_rate=sfreq)
      hrv_metrics = nk.hrv_time(rpeaks, sampling_rate=sfreq, show=False)
      features.append(hrv_metrics['HRV_MeanHR'].iloc[0])
      features.append(hrv_metrics['HRV_SDNN'].iloc[0])
    except:
      features.extend([0, 0])

    return np.array(features)






testEproch = splitRecordsOnDeltaT("/content/drive/MyDrive/data/265094/001/neurovisor_2025-04-05_10-51-45.edf", 5)
testEproch = testEproch.get_data();
paramVector = [extractParametrsMatrix (x) for x in testEproch]
print(len(paramVector[0]))






```

#Приложение 4
Обучение модели и проверка метрик


```python
import pathlib
import mne
import numpy as np
import scipy
import pandas as pd
from scipy.signal import welch
import neurokit2 as nk

from sklearn.model_selection import GroupShuffleSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


DATA_DIR = "/content/drive/MyDrive/data"
PARTICIPANTS_XLSX_PATH = "/content/drive/MyDrive/game_preferences_form_1.xlsx"

DT_SECONDS = 20
ECG_NAME = "ECG V3-V1"
FREQ_BANDS = {'delta': [0.5, 4], 'theta': [4, 8], 'alpha': [8, 13], 'beta': [13, 30], 'gamma': [30, 70]}


def getPathFile(fileDir: str, pattern: str) -> list:
    dir = pathlib.Path(fileDir)
    return [str(path) for path in dir.rglob(pattern)]

def getDict(base_dir: str) -> dict:
    tempDict = {}
    folder_paths = getPathFile(base_dir, "[0-9][0-9][0-9][0-9][0-9][0-9]")
    for path in folder_paths:
        c_name = pathlib.Path(path).name
        if c_name not in tempDict:
            edf_files = getPathFile(path, "*.edf")
            if edf_files:
                tempDict[c_name] = edf_files[0]
    return tempDict

def getInfoAboutSex(filePath: str) -> pd.DataFrame:
    data = pd.read_excel(filePath, dtype=str)
    uData = data[["ID", "Пол"]].copy()
    uData["Пол"] = uData["Пол"].map({"m": 0, "w": 1})
    uData["Пол"] = uData["Пол"].astype(int)
    return uData


def filterRecordComponent(eeg, ecg):
    if eeg:
      eeg.filter(l_freq=0.5, h_freq=70, verbose='WARNING')
      eeg.notch_filter(freqs=50, verbose='WARNING')
    if ecg: ecg.filter(l_freq=5, h_freq=35, verbose='WARNING')
    return eeg, ecg

def splitRecordOn(fileName):
    try:
        rawData = mne.io.read_raw_edf(fileName, preload=True, verbose="WARNING")
        dataEEG = rawData.copy()
        dataECG = None
        if ECG_NAME in dataEEG.ch_names:
            dataEEG.drop_channels([ECG_NAME])
            dataECG = rawData.copy().pick(ECG_NAME)
        return filterRecordComponent(dataEEG, dataECG)
    except Exception as e:
        return None, None

def splitRecordsOnDeltaT(fileName, dt):
    dataEEG, dataECG = splitRecordOn(fileName)
    if not dataEEG or not dataECG: return None

    combineFilterData = dataEEG.copy()
    combineFilterData.add_channels([dataECG], force_update_info=True)
    events = mne.make_fixed_length_events(combineFilterData, duration=dt)
    epochs = mne.Epochs(combineFilterData, events, tmin=0, tmax=dt, baseline=None, preload=True, verbose='WARNING')
    return epochs

def extractParametrsMatrix(epoch_matrix, sfreq=1000):
    eeg_data, ecg_data = epoch_matrix[:-1, :], epoch_matrix[-1, :]
    features = []
    for channel in eeg_data:
        freqs, psd = welch(channel, fs=sfreq, nperseg=int(sfreq))
        for band, (fmin, fmax) in FREQ_BANDS.items():
            features.append(np.sum(psd[np.logical_and(freqs >= fmin, freqs <= fmax)]))
        features.append(np.var(channel)) #Новые признаки для улучшения результатов модели
        features.append(scipy.stats.skew(channel))
        features.append(scipy.stats.kurtosis(channel))
    try:
        _, rpeaks = nk.ecg_peaks(ecg_data, sampling_rate=sfreq)
        hrv_metrics = nk.hrv_time(rpeaks, sampling_rate=sfreq, show=False)
        features.extend([hrv_metrics['HRV_MeanHR'].iloc[0], hrv_metrics['HRV_SDNN'].iloc[0]])
    except:
        features.extend([np.nan, np.nan])
    return np.array(features)



print("Шаг 1: Сбор путей к файлам и информации об участниках...")
file_paths_dict = getDict(DATA_DIR)
participants_df = getInfoAboutSex(PARTICIPANTS_XLSX_PATH)
print(f"Найдено {len(file_paths_dict)} файлов и {len(participants_df)} записей об участниках.")

print("\nШаг 2: Обработка файлов и извлечение признаков...")
all_X_features = []
all_y_labels = []
all_groups = []

for subject_id, file_path in file_paths_dict.items():
    print(f"--- Обработка ID: {subject_id} ---")

    # Ищем пол участника
    participant_info = participants_df[participants_df['ID'] == subject_id]
    if participant_info.empty:
        continue
    label = participant_info['Пол'].iloc[0]

    # Получаем окна
    epochs_object = splitRecordsOnDeltaT(file_path, DT_SECONDS)
    if epochs_object is None:
        continue

    # Извлекаем данные и признаки
    data_tensor = epochs_object.get_data()
    sfreq = epochs_object.info['sfreq']
    features = np.array([extractParametrsMatrix(epoch, sfreq) for epoch in data_tensor])

    # Добавляем результаты в общие списки
    all_X_features.append(features)
    all_y_labels.append(np.array([label] * len(features)))
    all_groups.append(np.array([subject_id] * len(features)))

print("\nШаг 3: Формирование итоговых датасетов...")
X = np.vstack(all_X_features)
y = np.concatenate(all_y_labels)
groups = np.concatenate(all_groups)

imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

print(f"Итоговая форма матрицы признаков X: {X.shape}")
print(f"Итоговая форма вектора меток y: {y.shape}")
print(f"Распределение классов: 0 (мужчины) - {np.sum(y==0)}, 1 (женщины) - {np.sum(y==1)}")


print("\nШаг 4: Разделение данных на обучающую и тестовую выборки...")
gss = GroupShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

from imblearn.over_sampling import SMOTE

print("\nШаг 4.5: Балансировка обучающей выборки с помощью SMOTE...")
smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("Распределение классов ДО SMOTE:")
print(pd.Series(y_train).value_counts())
print("\nРаспределение классов ПОСЛЕ SMOTE:")
print(pd.Series(y_train_resampled).value_counts())
print(f"Размер обучающей выборки: {X_train.shape[0]} эпох")
print(f"Размер тестовой выборки: {X_test.shape[0]} эпох")
print(f"Распределение классов в тесте: 0 - {np.sum(y_test==0)}, 1 - {np.sum(y_test==1)}")

print("\nШаг 5: Обучение модели RandomForest...")
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1)
model.fit(X_train_resampled, y_train_resampled)

print("\nШаг 6: Оценка модели на тестовых данных...")
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"\nТочность (Accuracy): {accuracy:.3f}")

print("\nОтчет по классификации:")
print(classification_report(y_test, y_pred, target_names=['Мужчина (0)', 'Женщина (1)']))

print("\nМатрица ошибок:")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Мужчина', 'Женщина'], yticklabels=['Мужчина', 'Женщина'])
plt.xlabel('Предсказанный класс')
plt.ylabel('Истинный класс')
plt.title(f'Матрица ошибок (dt = {DT_SECONDS}s)')
plt.show()
```
